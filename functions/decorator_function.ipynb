Decorator



@wraps(func)
def wrapper(...):
is just a shortcut for:

wrapper.__name__ = func.__name__
wrapper.__doc__ = func.__doc__
wrapper.__module__ = func.__module__
wrapper.__annotations__ = func.__annotations__
wrapper.__wrapped__ = func


```python
from functools import wraps


def decorator (func):
    @wraps(func)
    def wrapper(*args,**kwargs):
        print(f"Function {func.__name__} started")
        res = func(*args,**kwargs)
        print(f"Fucntion {func.__name__} ended")
        return  res
    return wrapper

@decorator
def crazy_function(a) :
    print(a)

crazy_function(10)
print(crazy_function.__name__)
```

Various Types of the Decorators

```python
# 4️⃣ Timing Decorator (NOW it will make sense)
from multiprocessing import process
import time

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"{func.__name__} took {time.time() - start:.2f}s")
        return result
    return wrapper

@timer
def solve(a):
    time.sleep(a)

solve(1)
solve(2)
```

```python
# 5️⃣ AUTH / PERMISSION Decorator (REAL BACKEND)


def user_logged_in () :
    return True


def login_required(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if not user_logged_in():
            raise Exception("Unauthorized")
        return func(*args, **kwargs)
    return wrapper

@login_required
def dashboard():
    print("Welcome")
    
    
dashboard()
```

RETRY Logic In Code

```python

def retries(times):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(3) :
                try:
                    return func(*args, **kwargs)
                except :
                    pass
            raise Exception(f"Failed After {times} retries ")
        return wrapper
    return decorator
                
```
