```python
# Practice

#✅ Q1 — Sum of squares of even numbers

nums = [1,2,3,4,5,6]

sosen =  sum(map(lambda x : x*x if x % 2 ==0 else 0,nums))
sosen =  sum(map(lambda x : x*x, filter(lambda x :  x %2 ==0,nums)))
sosen = sum(x*x for x in nums if  x%2 ==0)
print(sosen)
```
