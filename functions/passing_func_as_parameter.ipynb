```python
from functools import reduce


def add(*nums):                  #pack as tuple
    return sum(nums)             #iterate on tuple

def multiple(*nums):             #pack as tuple
    return reduce(lambda a,b : a*b, nums,1)    #iterate on tuple

def operator(func,*nums):          #collect as tuple
    return func(*nums)             #unpack tuple


print(operator(multiple,1,2,3,4,5))
print(operator(add,1,2,3,4,5))
```

```python
def is_even(a):
    return a&1 == 0

def is_odd(a):
    return a&1

def check(a,rule):
    return rule(a)

print(check(19,is_even))
print(check(19,is_odd))
```

```python
# 7️⃣ Passing lambda as argument (very common)

def operate(func,x):
    return func(x)

print(operate(lambda x : x*2, 100))
```
