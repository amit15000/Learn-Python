```python
#Reduce
# reduce(func,iterable, initialization)
# func must take two arguments: func(accumulator, current)

# initializer is optional but super useful (especially for empty lists)cc
```

```python
#Multiply all numbers
from functools import reduce

nums = [1,2,3,4]

res  = reduce(lambda a,b : a*b , nums)


nums =[]
res = reduce(lambda a,b : a*b , nums, 1)
print(res)
```
