```python
# Exercise 1 (Intermediate): Word Frequency Counter

sent = "I love python because python loves data"

record = {}

for ch in sent.lower():
    if ch == " ": continue
    
    if ch in record :
        record[ch]+= 1
    else :
        record[ch] = 1

print(record)
```

```python

sent = "I love python because python loves data"

record = {}


for ch in sent.lower():
    if ch != " ":
        record[ch] = record.get(ch,0)+1

print(record)
```

```python
from collections import Counter


sent = "I love python because python loves data"

record = Counter(sent.lower().replace(" ",""))
print(record)
```

```python
sent = "I love python because python loves data"

record = Counter(sent.lower().replace(" ",""))

print(record.most_common());
```
