```python
student = {
    "name": "Amit",
    "age": 23,
    "city": "Paris"
}

# print(student["name"])

print(student.get("name"))
print(student.get("salary"))  # None
print(student.get("salary",0)) #default
```

```python
# ðŸŸ¦ 4. Looping Through Dictionaries

for key in student :
    print(key,student[key], end=', ')
    
    
for key, value in student.items():
    print(key,value)
```

```python
# ðŸŸ¦ 6. Nested Dictionaries

student = {
    "name" : "Amit",
    "class" : {
        "section" : "D",
        "standard" : 12
    },
    "skills" : ['fronend','backend','ai-ml']
}

print(student["class"]["section"])
print(student.get('class').get('section'))
student["skills"].append("Hexa")
print(student.get("skills"))
```
