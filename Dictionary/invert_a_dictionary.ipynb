```python
#Nested Dictionary Inversion

data = {
    "student1": {"math": 90, "english": 80},
    "student2": {"math": 85, "english": 95},
    "student3": {"math": 90, "english": 70}
}


result = {}

for student, dicValue in data.items():
    for sub,mark in dicValue.items():
        result.setdefault(sub,{})
        result[sub].setdefault(mark,[])
        
        result[sub][mark].append(student)

print(result)
            
```

```python
#Nested Dictionary Inversion

data = {
    "student1": {"math": 90, "english": 80},
    "student2": {"math": 85, "english": 95},
    "student3": {"math": 90, "english": 70}
}


result = {}

for student, subMarks in data.items():
    for subject, marks in subMarks.items():
        
        if subject not in result: 
            result[subject] = {}
        if marks not in result[subject]:
            result[subject][marks]= []
        
        result[subject][marks].append(student)

print(result)
            
```

```python
data = {
    "u1": {
        "name": "Amit",
        "age": 20,
        "contact": {"email": "a@x.com", "phone": "111"},
        "skills": ["python", "sql"],
        "meta": {"city": "Delhi", "tags": ["founder", "ai"]}
    },
    "u2": {
        "name": "Sumit",
        "age": 23,
        "contact": {"email": "s@x.com", "phone": "111"},
        "skills": ["python", "ml"],
        "meta": {"city": "Delhi", "tags": ["ai"]}
    },
    "u3": {
        "name": "Amit",
        "age": 20,
        "contact": {"email": "a@x.com", "phone": "333"},
        "skills": [],
        "meta": {"city": "Mumbai", "tags": ["founder"]}
    }
}

result = {}

for userID, details in data.items():
    for title, value in details.items():
        
        if title in ("name", "age"):
            if title not in result:
                result[title] = {}
            if value not in result[title]:
                result[title][value] = []
            result[title][value].append(userID)
                    
        
        if title == "contact":
            for emailorphone, epvalue in value.items():
                newTitle = f"contact.{emailorphone}"
                if newTitle not in result:
                    result[newTitle] = {}
                if epvalue not in result[newTitle]:
                    result[newTitle][epvalue] = []
                if userID not in result[newTitle][epvalue]:
                    result[newTitle][epvalue].append(userID)
        
        
        if title == "skills":
            newTitle = "skills[]"
            if newTitle not in result:
                result[newTitle] = {}
            
            for skill in value:
                if skill not in result[newTitle]:
                    result[newTitle][skill] = []
                result[newTitle][skill].append(userID)
        
        
        if title == "meta":
            for cityOrTags, ctvalue in value.items():
                newTitle = "meta.city" if cityOrTags == "city" else "meta.tags[]"
                
                if(newTitle == "meta.city"):
                    if newTitle not in result:
                        result[newTitle] = {}
                    if ctvalue not in result[newTitle] :
                        result[newTitle][ctvalue] = []
                    result[newTitle][ctvalue].append(userID)
                
                else :       #means we are on tags
                    if newTitle not in result:
                        result[newTitle] = {}
                    
                    for tags in ctvalue:
                        if tags not in result[newTitle]:
                            result[newTitle][tags] = []
                        result[newTitle][tags].append(userID)                                               
                    
print(result)
```

```python
def invert_nested(data: dict) -> dict:
    result = {}

    def add(path: str, value, user_id: str):
        # Create nested dicts safely
        if path not in result:
            result[path] = {}

        if value not in result[path]:
            result[path][value] = []

        # Avoid duplicates
        if user_id not in result[path][value]:
            result[path][value].append(user_id)

    def walk(obj, user_id: str, path: str = ""):
        # obj can be dict, list, or scalar
        if obj is None:
            return

        if isinstance(obj, dict):
            for k, v in obj.items():
                new_path = f"{path}.{k}" if path else k
                walk(v, user_id, new_path)

        elif isinstance(obj, list):
            # For lists, we index each item at "path[]"
            list_path = f"{path}[]"
            for item in obj:
                if item is None:
                    continue
                if isinstance(item, (dict, list)):
                    # If list contains dicts/lists, keep walking deeper
                    walk(item, user_id, list_path)
                else:
                    add(list_path, item, user_id)

        else:
            # scalar value
            add(path, obj, user_id)

    # Build index
    for user_id, profile in data.items():
        walk(profile, user_id)

    # Sort user lists for stable output
    for path in result:
        for value in result[path]:
            result[path][value].sort()

    return result


# ---- test on your data ----
data = {
    "u1": {
        "name": "Amit",
        "age": 20,
        "contact": {"email": "a@x.com", "phone": "111"},
        "skills": ["python", "sql"],
        "meta": {"city": "Delhi", "tags": ["founder", "ai"]}
    },
    "u2": {
        "name": "Sumit",
        "age": 23,
        "contact": {"email": "s@x.com", "phone": "111"},
        "skills": ["python", "ml"],
        "meta": {"city": "Delhi", "tags": ["ai"]}
    },
    "u3": {
        "name": "Amit",
        "age": 20,
        "contact": {"email": "a@x.com", "phone": "333"},
        "skills": [],
        "meta": {"city": "Mumbai", "tags": ["founder"]}
    }
}

result = invert_nested(data)
print(result)
```
