```python
# Nested Dictionary Parser

user = {
    "id": 101,
    "profile": {
        "name": "Amit",
        "education": {
            "degree": "MSc",
            "year": 2025
        }
    }
}




def parser(obj,path:str=""):
    
    if isinstance(obj, dict):
        for k,v in obj.items():
            newPath = f"{path}.{k}" if path else k
            
            parser(v,newPath)
    else :
        
        print(f"{path} -> {obj}")

parser(user)
    
```
